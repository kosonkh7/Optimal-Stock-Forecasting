## 배송일자 + 수하인 구 기준으로 SUM 그룹핑

```python
sidoSeoulgu = pd.DataFrame(columns=['배송년월일', '수하인_구명', '수하인_구코드', '대분류_착지물동량 가구/인테리어', '대분류_착지물동량 기타',
       '대분류_착지물동량 도서/음반', '대분류_착지물동량 디지털/가전', '대분류_착지물동량 생활/건강',
       '대분류_착지물동량 스포츠/레저', '대분류_착지물동량 식품', '대분류_착지물동량 출산/육아', '대분류_착지물동량 패션의류',
       '대분류_착지물동량 패션잡화', '대분류_착지물동량 화장품/미용'])

for data_dir in data_list:
    data = pd.read_csv(path + '/sidoSeoulgu/' + data_dir, encoding="euc-kr")
    data = data.groupby(['배송년월일', '수하인_구명', '수하인_구코드'])[['대분류_착지물동량 가구/인테리어', '대분류_착지물동량 기타',
           '대분류_착지물동량 도서/음반', '대분류_착지물동량 디지털/가전', '대분류_착지물동량 생활/건강',
           '대분류_착지물동량 스포츠/레저', '대분류_착지물동량 식품', '대분류_착지물동량 출산/육아', '대분류_착지물동량 패션의류',
           '대분류_착지물동량 패션잡화', '대분류_착지물동량 화장품/미용']].sum().reset_index()
    sidoSeoulgu = pd.concat([sidoSeoulgu, data])
    del data

sidoSeoulgu.reset_index(drop=True, inplace=True)
sidoSeoulgu = sidoSeoulgu.astype(int, errors='ignore')
```

# 결측치 제거하여 병합
특정 배송일자-수하구에 결측치가 존재한다.
이를 0으로 채워주고 seoulGuGu과 더하는 과정

```python
seoulGuGu = pd.read_csv("seoulGuGu.csv", encoding='euc-kr')
```

```python
# 1. 수하인 구명과 구코드 조합
recipient_data = {
    '수하인_구명': ['강남구', '송파구', '영등포구', '강서구', '서초구', '성북구', '도봉구', '용산구', 
                  '관악구', '강동구', '양천구', '중구', '성동구', '동작구', '은평구', '노원구', 
                  '금천구', '구로구', '광진구', '동대문구', '마포구', '서대문구', '강북구', '중랑구', '종로구'],
    '수하인_구코드': [11680, 11710, 11560, 11500, 11650, 11290, 11320, 11170, 
                  11620, 11740, 11470, 11140, 11200, 11590, 11380, 11350, 
                  11545, 11530, 11215, 11230, 11440, 11410, 11305, 11260, 11110]
}
recipient_df = pd.DataFrame(recipient_data)

# 2. 배송년월일 생성
dates = pd.date_range(start='2018-01-01', end='2023-12-31', freq='D')
dates_df = pd.DataFrame({'배송년월일': dates.strftime('%Y%m%d')})

# 3. 모든 조합 생성 (수하인_구명 & 수하인_구코드는 고정된 세트)
all_combinations = recipient_df.merge(dates_df, how='cross')

# 4. 나머지 컬럼 0으로 초기화
columns = ['대분류_착지물동량 가구/인테리어', '대분류_착지물동량 기타', '대분류_착지물동량 도서/음반',
           '대분류_착지물동량 디지털/가전', '대분류_착지물동량 생활/건강', '대분류_착지물동량 스포츠/레저',
           '대분류_착지물동량 식품', '대분류_착지물동량 출산/육아', '대분류_착지물동량 패션의류',
           '대분류_착지물동량 패션잡화', '대분류_착지물동량 화장품/미용']
for col in columns:
    all_combinations[col] = 0
```

```python
all_combinations['배송년월일'] = all_combinations['배송년월일'].astype('int64')
```

```python
all_combinations.info()
```

```python
# 5. df1 데이터 반영
# df1_grouped = df1.groupby(['배송년월일', '수하인_구명', '수하인_구코드']).sum().reset_index()
all_combinations = all_combinations.merge(seoulGuGu, on=['배송년월일', '수하인_구명', '수하인_구코드'], how='left', suffixes=('', '_df1'))
for col in columns:
    all_combinations[col] = all_combinations[col] + all_combinations[f"{col}_df1"].fillna(0)
all_combinations = all_combinations.drop(columns=[f"{col}_df1" for col in columns])

# 6. df2 데이터 반영
# df2_grouped = df2.groupby(['배송년월일', '수하인_구명', '수하인_구코드']).sum().reset_index()
all_combinations = all_combinations.merge(sidoSeoulgu, on=['배송년월일', '수하인_구명', '수하인_구코드'], how='left', suffixes=('', '_df2'))
for col in columns:
    all_combinations[col] = all_combinations[col] + all_combinations[f"{col}_df2"].fillna(0)
all_combinations = all_combinations.drop(columns=[f"{col}_df2" for col in columns])
```
